```python
from sklearn.preprocessing import StandardScaler
```

```python
X = StandardScaler().fit_transform(df[["Time on App","Length of Membership"]])
```

```python
X[:5]
```

Gradient descent is sensitive to scale.

Features on similar scales lead to faster and more stable convergence.

Prevents dominance of features with larger ranges.

```python
X_bias = np.c_[np.ones((X.shape[0], 1)), X] 
```

```python
y = df['Yearly Amount Spent'].values.reshape(-1, 1)
```

```python
def gradient_descent(X,y,lr=0.01,iterations=1000):
    m = len(y)
    theta = np.zeros((X.shape[1], 1))
    cost_history = []
    for _ in range(iterations):
        y_pred = X.dot(theta)
        cost  = (1/(2*m)) * np.sum(np.square(y_pred - y))
        cost_history.append(cost)
        gradient = (2/m) * X.T.dot(y_pred - y)
        theta -= lr * gradient
    return theta, cost_history
```

```python
theta_gd, cost_history = gradient_descent(X_bias,y)
```

```python
print("Weights from Gradient Descent:", theta_gd.ravel())
```

```python
import plotly.graph_objects as go
```

```python
figure = go.Figure()
```

```python
figure.add_trace(go.Scatter3d(
    x=X[:, 0],
    y=X[:, 1],
    z=y.flatten(),
    mode='markers',
    marker=dict(size=3, color='blue'),
    name='Actual'
))
figure.add_trace(go.Mesh3d(
    x=X[:, 0],
    y=X[:, 1],
    z=X_bias.dot(theta_gd).flatten(),
    opacity=0.5,
    color='red',
    name='Prediction Surface'
))

figure.update_layout(
    title='3D Linear Regression Surface (Plotly)',
    scene=dict(
        xaxis_title='Time on App (scaled)',
        yaxis_title='Length of Membership (scaled)',
        zaxis_title='Yearly Amount Spent'
    )
)
figure.update_layout(width=1200, height=900)
```

```python
fig_curve = px.line(
    y=cost_history,
    labels={'x': 'Iterations', 'y': 'Cost'},
    title='Gradient Descent Convergence (Plotly)'
)
```

```python
fig_curve.show()
```

```python
X_normal = np.c_[np.ones((X.shape[0], 1)), X]
```

```python
theta_ls = np.linalg.inv(X_normal.T.dot(X_normal)).dot(X_normal.T).dot(y)
```

```python
y_pred_gd = X_bias.dot(theta_gd)
```

```python
y_pred_ls = X_bias.dot(theta_ls)
```

```python
fig_prediction = go.Figure()

fig_prediction.add_trace(go.Scatter(
    x=list(range(len(y))),
    y=y.flatten(),
    mode='lines',
    name='Actual',
    line=dict(color='blue')
))

fig_prediction.add_trace(go.Scatter(
    x=list(range(len(y))),
    y=y_pred_gd.flatten(),
    mode='lines',
    name='Gradient Descent',
    line=dict(color='red')
))

fig_prediction.add_trace(go.Scatter(
    x=list(range(len(y))),
    y=y_pred_ls.flatten(),
    mode='lines',
    name='Least Squares',
    line=dict(color='green', dash='dash')
))

fig_prediction.update_layout(
    title='Prediction Comparison',
    xaxis_title='Sample Index',
    yaxis_title='Yearly Amount Spent',
    legend=dict(x=0.01, y=0.99)
)

fig_prediction.show()
```
