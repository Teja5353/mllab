```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sklearn.datasets as data
```

```python
iris = data.load_iris()
```

```python
df = pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
df["target"] = iris.target
df["species"] = pd.Categorical.from_codes(iris.target,iris.target_names)
```

```python
df = df[(df["target"] == 0) | (df["target"] == 1)]
```

```python
import seaborn as sns

sns.pairplot(df, hue='species')
plt.suptitle('Pairplot of Features (Setosa vs. Versicolor)', y=1.02)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[iris.feature_names].values
y = df["target"].values.reshape(-1,1)
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=42,test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def compute_cost(X, y, weights):
    m = len(y)
    h = sigmoid(X @ weights)
    epsilon = 1e-5
    cost = -1/m * (y.T @ np.log(h + epsilon) + (1 - y).T @ np.log(1 - h + epsilon))
    return cost[0][0]
```

```python
def logistic_regression(X, y, lr=0.01, iterations=1000):
    m, n = X.shape
    X_b = np.c_[np.ones((m, 1)), X]
    weights = np.zeros((n + 1, 1))
    cost_history = []

    for i in range(iterations):
        h = sigmoid(X_b @ weights)
        gradient = X_b.T @ (h - y) / m
        weights -= lr * gradient
        cost_history.append(compute_cost(X_b, y, weights))

    return weights, cost_history
```

```python
weights, cost_history = logistic_regression(X_train,y_train,lr=0.1,iterations=1000)
```

```python
plt.plot(cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Logistic Regression Cost Convergence')
plt.grid(True)
plt.show()
```

```python
def predict(X, weights):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    probs = sigmoid(X_b @ weights)
    return (probs >= 0.5).astype(int)
```

```python
y_pred = predict(X_test, weights)
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_curve, auc
)
```

```python
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {acc:.2f}")
print(f"Precision: {prec:.2f}")
print(f"Recall: {rec:.2f}")
print(f"F1-score: {f1:.2f}")
```

```python
y_probs = sigmoid(np.c_[np.ones((X_test.shape[0], 1)), X_test] @ weights)
fpr, tpr, thresholds = roc_curve(y_test, y_probs)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
